```python
import pandas as pd
```

```python
comments = pd.read_parquet(
    "FattoSentimentComments.parquet.snappy", engine="fastparquet"
)
```

```python
articles = pd.read_parquet(
    "dailymailWithSentiment.parquet.snappy", engine="fastparquet"
)
```

```python
comments.head()
```

```python
articles.head()
```

```python
ds = (
    comments.merge(
        articles, how="right", on="link", suffixes=("_Comments", "_Articles")
    )
    .drop(axis=1, labels="comment")
    .drop_duplicates(subset="link")
)
```

```python
ds.head()
```

```python
ds.to_parquet("dailySentimentComplete.parquet.snappy", engine="fastparquet")
```
