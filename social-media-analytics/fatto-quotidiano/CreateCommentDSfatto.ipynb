```python
from itertools import islice

import pandas as pd
```

```python
ds = pd.read_parquet("fatto.parquet.snappy", engine="fastparquet")
```

```python
ds.head()
```

```python
comments = pd.DataFrame(
    {"link": pd.Series([], dtype="string"), "comment": pd.Series([], dtype="string"),}
)
```

```python
# Creating an empty list to store dictionaries
comment_list = []

# Iterating over rows
for _, row in ds.iterrows():
    # Iterating over comments using islice
    for comment in islice(row["comments"], 10):
        # Creating a dictionary for each row
        comment_dict = {"link": row["link"], "comment": comment}
        # Appending the dictionary to the list
        comment_list.append(comment_dict)

# Creating DataFrame from the list of dictionaries
comments = pd.DataFrame(comment_list)
```

```python
comments.head()
```

```python
comments.to_parquet("fattoComments.parquet.snappy", engine="fastparquet")
```
